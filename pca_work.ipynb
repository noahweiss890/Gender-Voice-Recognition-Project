```python
import json
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

# 208 features

```python
data=json.loads(open("data/males_females_audio.json").read())

classes=list(data)
features=list()
labels=list()
for i in range(len(classes)):
    for j in range(len(data[classes[i]])):
        feature=data[classes[i]][j][:208]
        features.append(feature)
        labels.append(classes[i])

scaler = StandardScaler()
data_rescaled = scaler.fit_transform(features)

pca = PCA().fit(data_rescaled)
plt.style.use("ggplot") 
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.title("208 features")
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')

derivatives = []
for i in range(1, len(np.cumsum(pca.explained_variance_ratio_))):
    y1 = np.cumsum(pca.explained_variance_ratio_)[i - 1]
    y2 = np.cumsum(pca.explained_variance_ratio_)[i]
    x1 = i - 1
    x2 = i
    derivative = (y2 - y1) / (x2 - x1)
    derivatives.append(derivative)
```

```python
def find_closest_to_zero_208(derivatives):
  min_diff = 0.0005 #good for 208
  min_index = None
  for i, d in enumerate(derivatives):
    diff = abs(d)
    if diff < min_diff:
        min_index = i
        break
  return derivatives[min_index], min_index


find_closest_to_zero_208(derivatives)
```
